```python
print("--------SECTION-6---------")
print("INFERENTIAL STATISTICS")

from scipy import stats

male = df[df['Gender'] == 'Male']['Spending Score (1-100)']
female = df[df['Gender'] == 'Female']['Spending Score (1-100)']

t_stat, p_value = stats.ttest_ind(male, female)

print("t-statistic =", t_stat)
print("p-value =", p_value)
```

What does p-value tell you?

It measures the evidence against the null hypothesis.

Is the difference significant?

No, because p-value (0.422) is greater than 0.05.

```python
print("--------SECTION-7---------")
print("CONFIDENCE INTERVAL")

import scipy.stats as stats

spending = df['Spending Score (1-100)']

mean = np.mean(spending)
std = np.std(spending, ddof=1)
n = len(spending)

margin = 1.96 * (std / np.sqrt(n))

lower = mean - margin
upper = mean + margin

print("95% CI:", lower, upper)
```

What range do customers fall into?

Average spending score likely lies between: 46.62 and 53.78

How confident are we?

We are 95% confident that the true population mean spending score falls within this interval.
